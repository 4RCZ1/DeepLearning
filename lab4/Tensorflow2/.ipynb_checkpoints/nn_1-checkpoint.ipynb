##  1.1. Implementacja algorytmu perceptronu 

\begin{itemize}
\item \textbf{__init__()} - ustawienie parametrów algorytmu (szybkość uczenia, ilość epok, losowa/nielosowa inicjalizacja wag)
\item \textbf{fit(dane, etykiety)}: 
\begin{itemize}
\item pobranie danych (dwie klasy - dane są utworzone przez dwie wybrane  klasy z iris, gdyż perceptron jest klasyfikatorem binarnym), 
\item pobranie etykiet {+1,-1}, 
\item inicjalizacja wag, 
\item inicjalizacja wektora błędu (ze względu na formułę obliczenia (zmienna update) jest to rodzaj funkcji straty, a nie funkcji dokładności), 
\item pętla po ilości iteracji: w każdej iteracji wszystkie dane, w każdej iteracji obliczenie sumy błędów na wszystkich danych i dodanie tej wartości jako kolejnego elementu wektora błędu -  w ten spośób wektor błędu przechowuje całą historię jakości uczenia
\end{itemize}
\end{itemize}

```python
class Perceptron(object):
    """Klasyfikator binarny

    Parametery
    ------------
    eta : float
      Szybkość uczenia (pomiędzy 0.0 i 1.0)
    n_iter : int
      ilość epok
    random_state : int
      losowa wartość dla generatora wag 

    Atrybuty
    -----------
    w_ : 1d-array
      wagi po procesie uczenia
    errors_ : list
      liczba złych klasyfikacji w danej epoce

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """Funkcja realizująca proces uczenia

        Parametry
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Wektory treningowe, gdzie n_examples jest liczbą przykładów, a
          n_features jest liczbą cech.
        y : array-like, shape = [n_examples]
          Wektor wartości do procesu uczenia (etykiety klas).

        Wartość zwracana
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.errors_ = []

        # pętla po ilości epok
        for _ in range(self.n_iter):
            #pętla dla jednej epoki
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

### 1.2.6. Prezentacja modelu dla nowych danych 

```python


def plot_decision_regions(X, y, classifier, resolution=0.01):

    markers = ('s', 'o', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    print(cmap)
    
    # granice wykresu
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    #utworzenie siatki punktów dla obliczeń
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    print(xx1.ravel())
    
    # obliczenia modelu na siatce - wynikiem jest wektor
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    #print(Z)
    
    #obliczone wartości jako siatka
    Z = Z.reshape(xx1.shape)
    #print(Z)
    
            
    print(Z)
    plt.contourf(xx1, xx2, Z, alpha=0.5, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class examples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
```

```python
plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')


# plt.savefig('images/02_08.png', dpi=300)
plt.show()
```

## 1.4 Adaptacyjne neurony liniowe i zbieżność uczenia

Schemat metody największego spadku w algorytmie Adaline (Adaptacyjny liniowy neuron)  jest opisany krótko w sekcji 2.2  na stronie [wstęp do optymalizacji](optymalizacja1.ipynb). W tym przypadku funkcja aktywacji jest funkcją identycznościową ($\varphi(z)=z$). Ponadto inaczej niż w podstawowym algorytmie perceptronu wagi są aktualizowane na całym zbiorze próbek treningowych (nie jest to algorytm typu stochastycznego, tylko klasyczna metoda największego spadku ze stałą szybkością uczenia - w klasyfikacji metod spadkowych nosi ona nazwę \textit{ metody gradientu prostego}). W nomenklaturze sieci neuronowych szybkość uczenia i ilość epok są \textit{hiperparametrami} algorytmu. Przyjmując stałą szybkość uczenia zwykle trzeba wykonać testy pozwalające określić w miarę dobrą jej wartosć.

## 1.4.1 Implementacja algorytmu Adaline w Pythonie

```python
class AdalineGD(object):
    """Klasyfikator — ADAptacyjny LIniowy NEuron.

    Parametry
    ------------
    eta : zmiennoprzecinkowy
        Współczynnik uczenia (w zakresie pomiędzy 0.0 i 1.0)
    n_iter : liczba całkowita
        Liczba przebiegów po zestawie uczącym.
    random_state : liczba całkowita
      Ziarno generatora liczb losowych służący do losowego
      inicjalizowania wag.


    Atrybuty
    -----------
    w_ : jednowymiarowa tablica
        Wagi po dopasowaniu.
    cost_ : lista
      Suma kwadratów błędów (wartość funkcji kosztu) w każdej epoce.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """ Trenowanie za pomocą danych uczących.

        Parametry
        ----------
        X : {tablicopodobny}, wymiary = [n_przykładów, n_cech]
            Wektory uczenia,
            gdzie n_przykładów oznacza liczbę przykładów, a
            n_cech – liczbę cech.
        y : tablicopodobny, wymiary = [n_przykładów]
            Wartości docelowe.

        Zwraca
        -------
        self : obiekt

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            net_input = self.net_input(X)
            # Zwróć uwagę, że metoda "activation" nie ma wpływu na działanie kodu,
            # gdyż jest to zwykła funkcja tożsamościowa. Moglibyśmy zamiast tego napisać
            # bezpośrednio po prostu `output = self.net_input(X)`.
            # Zastosowanie tej aktywacji w tym miejscu zapewnia jednolitość spojrzenia na algorytmikę sieci neuronowych
            output = self.activation(net_input)
            errors = (y - output)
            #print(y)
            #print(output)
            #print(errors)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            #print(cost)
            self.cost_.append(cost)
            #print(self.cost_)
        return self

    def net_input(self, X):
        """oblicza pobudzenie całkowite"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        """oblicza liniową funkcję aktywacji"""
        return X

    def predict(self, X):
        """zwraca etykietę klas po wykonaniu skoku jednostkowego"""
        return np.where(self.activation(self.net_input(X)) >= 0.0, 1, -1)
```

### 1.4.1.1 Test algorytmu dla dwóch istotnie różnych szybkości uczenia

Jak widać na wykresach zaprezentowanych poniżej, natrafiamy na dwa różne problemy.
Wykres po lewej przedstawia sytuację, gdy dobieramy zbyt dużą wartość współczynnika
uczenia. Zamiast minimalizacji funkcji kosztu następuje powiększanie błędu wraz
z każdą epoką, ponieważ przeskakujemy nad minimum globalnym. Z drugiej strony widzimy,

## 1.4.2. Usprawnianie gradientu prostego poprzez skalowanie cech

Metoda gradientu prostego stanowi jeden z wielu algorytmów, w których przydatne okazuje
się tzw. \textit{skalowanie cech}. Zastosujemy tu metodę skalowania zwaną \textit{ standaryzacją},
gdyż potraktowane nią dane uzyskują własności standardowego rozkładu normalnego: średnią
o wartości zero i wariancję jednostkową. Taka procedura normalizacji przyśpiesza proces
uzyskiwania zbieżności; nie sprawia to jednak, że pierwotny zestaw danych uzyskuje rozkład
normalny - standaryzację wykonujemy na każdej cesze oddzielnie. Standaryzacja przesuwa średnią każdej cechy w taki sposób, że zostaje wyśrodkowana do wartości 0, a każda cecha zawiera odchylenie standardowe równe 1 (wariancja jednostkowa):
$$
x_j^{'}=\frac{x_j-\mu_j}{\sigma_j}
$$
Tutaj $x_j$ jest wektorem składającym się z wartości j-tej cechy wszystkich przykładów uczących, a $\mu_j$ i $\sigma_j$ oznaczają odpowiednio średnią i wariancję tej cechy. Standaryzacja pomaga w nauce gradientu prostego m.in. dlatego, że funkcja optymalizująca musi wykonać mniej przebiegów w celu znalezienia optymalnego rozwiązania (globalnego
minimum kosztu), co zostało zaprezentowane na poniższym rysunku; obydwa wykresy ukazują
powierzchnię kosztu jako funkcję dwóch przykładowych wag w zagadnieniu klasyfikacji
dwuwymiarowej.

### 1.3.2.1. Standaryzacja dwóch pierwszych cech 

```python
# standaryzuje cechy
X_std = np.copy(X)
X_std[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
X_std[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
print(X[:,0])
print(X_std[:,0])
```

### 1.4.2.2. Uczenie  wystandaryzowanego modelu Adaline dla poprzednio rozbieżnej szybkości uczenia

Po wprowadzeniu standaryzacji ponownie wyuczymy model Adaline i sprawdzamy, czy algorytm będzie zbieżny po wykonaniu 
niewielkiej liczby przebiegów przy współczynniku uczenia o wartości $η = 0.01$. Jak widać, algorytm Adaline stał się zbieżny po uczeniu się na standaryzowanych cechach. Zauważ jednak, że suma kwadratów błędów
pozostaje niezerowa pomimo właściwego sklasyfikowania wszystkich przykładów kwiatów.

```python
ada_gd = AdalineGD(n_iter=15, eta=0.01)
ada_gd.fit(X_std, y)

plot_decision_regions(X_std, y, classifier=ada_gd)
plt.title('Adaline - gradient prosty')
plt.xlabel('Długość działki [standaryzowana]')
plt.ylabel('Długość płatka [standaryzowana]')
plt.legend(loc='upper left')
plt.tight_layout()
# plt.savefig('rysunki/02_14_1.png', dpi=300)
plt.show()

plt.plot(range(1, len(ada_gd.cost_) + 1), ada_gd.cost_, marker='o')
plt.xlabel('Epoki')
plt.ylabel('Suma kwadratów błędów')

plt.tight_layout()
# plt.savefig('rysunki/02_14_2.png', dpi=300)
plt.show()
```

## 1.4.3. Wielkoskalowe uczenie maszynowe i metoda stochastycznego spadku wzdłuż gradientu

W przypadku bardzo dużego zbioru próbek treningowych stosowanie metody gradientu prostego bywa dość kosztowne pod względem obliczeniowym, ponieważ musimy od nowa oceniać cały zbiór danych uczących za każdym razem, gdy wykonujemy
kolejny krok w kierunku globalnego minimum (Algorytm taki czasami nazywamy jest algorytmem \textit{wsadowym} gradientu prostego, bo pobiera w każdym kroku cały zbiór próbek).

Popularnym zamiennikiem algorytmu wsadowego gradientu prostego jest \textit{metoda stochastycznego spadku wzdłuż gradientu} (ang. stochastic gradient descent — SGD), czasami nazywana
także \textit{ iteracyjnym algorytmem spadku wzdłuż gradientu}. W tym przypadku w danym kroku poprawy wektora wag używamy jedynie pojedynczej próbki.

Algorytm SGD zazwyczaj
umożliwia  znacznie szybsze uzyskanie zbieżności, gdyż wagi są częściej aktualizowane.
Każdy gradient jest wyliczany na podstawie pojedynczej próbki uczącej, dlatego powierzchnia
błędów generuje większe szumy niż w gradiencie prostym, co również ma znaczenie, gdyż
dzięki temu algorytm SGD łatwiej może ignorować płytkie minima lokalne w przypadku,
gdy pracujemy z nieliniowymi funkcjami kosztu. Aby uzyskać satysfakcjonujące
wyniki za pomocą algorytmu SGD, bardzo ważne jest zaprezentowanie algorytmowi
danych uczących w przypadkowej kolejności, a ponadto  przed każdą epoką przetasować
zestaw danych uczących, aby uniknąć cykliczności.

W implementacjach stochastycznego spadku wzdłuż gradientu niezmienna stała szybkości uczenia $η$
często jest zastępowany adaptacyjnym współczynnikiem uczenia, którego wartość maleje wraz z upływem
czasu; może on przybrać np. następującą postać:
$$
\frac{c_1}{(liczba\ iteracji) + c_2}
$$
gdzie $c_1,\ c_2$ są stałymi. Dzięki adaptacyjnemu współczynnikowi uczenia możemy jeszcze bardziej
zbliżyć się do minimum kosztu.

Kolejną zaletą metody SGD jest możliwość wykorzystania jej do \textit{ uczenia przyrostowego}.
Rozwiązanie to polega na trenowaniu modelu za pomocą ciągle napływających nowych danych
uczących. Jest to technika przydatna zwłaszcza w sytuacji gromadzenia dużych ilości
informacji. Poprzez uczenie w locie system jest w stanie natychmiastowo dostosować się do zmian.


### 1.4.3.1. Implementacja algorytmu Adaline w wersji stochastycznej

Teraz wewnątrz metody \textit{fit} będziemy aktualizować wagi po
sprawdzeniu każdego przykładu uczącego.

Następnie zaimplementujemy dodatkową metodę
\textit{partial\_fit}, która nie inicjuje od nowa wag w przypadku uczenia w locie. 

Aby sprawdzić zbieżność algorytmu po treningu, będziemy obliczać koszt jako średni koszt przykładów
uczących w każdej epoce. 

Ponadto dodamy możliwość tasowania (shuffle) danych uczących
przed rozpoczęciem każdej epoki, dzięki czemu unikniemy cykliczności podczas optymalizowania
funkcji kosztu. 

Stosowana w klasyfikatorze AdalineSGD metoda \_shuffle działa w następujący sposób: dzięki
funkcji permutation w np.random generujemy losową sekwencję unikatowych liczb w przedziale
od 0 do 100. Liczby te są następnie wykorzystywane jako indeksy umożliwiające tasowanie
macierzy cech i wektora etykiet klas.


```python
class AdalineSGD(object):
    """Klasyfikator — ADAptacyjny LIniowy NEuron.

    Parametry
    ------------
    eta : zmiennoprzecinkowy
        Współczynnik uczenia (w zakresie pomiędzy 0.0 i 1.0)
    n_iter : liczba całkowita
        Liczba przebiegów po zestawie uczącym.
    shuffle : wartość boolowska (domyślnie: True)
        Jeżeli jest ustalona wartość True,
        tasuje dane uczące przed każdą epoką w celu zapobiegnięcia cykliczności.
    random_state : liczba całkowita
      Ziarno generatora liczb losowych służący do losowego
      inicjalizowania wag.


    Atrybuty
    -----------
    w_ : jednowymiarowa tablica
        Wagi po dopasowaniu.
    cost_ : lista
      Suma kwadratów błędów (funkcja kosztu) uśredniona
      wobec wszystkich przykładów uczących w każdej epoce.

        
    """
    def __init__(self, eta=0.01, n_iter=10, shuffle=True, random_state=None):
        self.eta = eta
        self.n_iter = n_iter
        self.w_initialized = False
        self.shuffle = shuffle
        self.random_state = random_state
        
    def fit(self, X, y):
        """ Dopasowanie danych uczących.

        Parametry
        ----------
        X : {tablicopodobny}, wymiary = [n_przykładów, n_cech]
            Wektory uczące, gdzie n_przykładów
            oznacza liczbę przykładów, a
            n_cech określa liczbę cech.
        y : tablicopodobny, wymiary = [n_przykładów]
            Wartości docelowe.

        Zwraca
        -------
        self : obiekt

        """
        self._initialize_weights(X.shape[1])
        self.cost_ = []
        for i in range(self.n_iter):
            if self.shuffle:
                X, y = self._shuffle(X, y)
            cost = []
            for xi, target in zip(X, y):
                cost.append(self._update_weights(xi, target))
            avg_cost = sum(cost) / len(y)
            self.cost_.append(avg_cost)
        return self

    def partial_fit(self, X, y):
        """dopasowuje dane uczące bez ponownej inicjalizacji wag"""
        if not self.w_initialized:
            self._initialize_weights(X.shape[1])
        if y.ravel().shape[0] > 1:
            for xi, target in zip(X, y):
                self._update_weights(xi, target)
        else:
            self._update_weights(X, y)
        return self

    def _shuffle(self, X, y):
        """tasuje dane uczące"""
        r = self.rgen.permutation(len(y))
        return X[r], y[r]
    
    def _initialize_weights(self, m):
        """inicjalizuje wagi z bardzo małymi wartościami"""
        self.rgen = np.random.RandomState(self.random_state)
        self.w_ = self.rgen.normal(loc=0.0, scale=0.01, size=1 + m)
        self.w_initialized = True
        
    def _update_weights(self, xi, target):
        """wykorzystuje regułę uczenia Adaline do aktualizacji wag"""
        output = self.activation(self.net_input(xi))
        error = (target - output)
        self.w_[1:] += self.eta * xi.dot(error)
        self.w_[0] += self.eta * error
        cost = 0.5 * error**2
        return cost
    
    def net_input(self, X):
        """oblicza pobudzenie całkowite"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        """oblicza liniową funkcję aktywacji"""
        return X

    def predict(self, X):
        """zwraca etykietę klas po wykonaniu skoku jednostkowego"""
        return np.where(self.activation(self.net_input(X)) >= 0.0, 1, -1)
```

```python
ada_sgd = AdalineSGD(n_iter=15, eta=0.01, random_state=1)
ada_sgd.fit(X_std, y)

plot_decision_regions(X_std, y, classifier=ada_sgd)
plt.title('Adaline - stochastyczny spadek wzdłuż gradientu')
plt.xlabel('Długość działki [standaryzowana]')
plt.ylabel('Długość płatka [standaryzowana]')
plt.legend(loc='upper left')

plt.tight_layout()
# plt.savefig('rysunki/02_15_1.png', dpi=300)
plt.show()

plt.plot(range(1, len(ada_sgd.cost_) + 1), ada_sgd.cost_, marker='o')
plt.xlabel('Epoki')
plt.ylabel('Średni koszt')

plt.tight_layout()
# plt.savefig('rysunki/02_15_2.png', dpi=300)
plt.show()
```

#  2. Przetwarzanie danych kategorialnych


Rzeczywiste zestawy danych zawierają często przynajmniej jedną kolumnę cech kategorialnych (nienumerycznych) - zwykle  etykiety klas są tego typu. Ponadto wiele algorytmów uczenia maszynowego wymaga, aby etykiety klas były kodowane w postaci liczb całkowitych. 


## 2.1. Cechy nominalne i porządkowe

Dane kategorialne dzielimy na dwie grupy:  cechy \textit{nominalne}
i \textit{porządkowe}. 
\begin{enumerate}
\item Cechy porządkowe to takie wartości kategoryzujące, które możemy wykorzystać
do sortowania lub porządkowania informacji. Przykładowo rozmiar koszulki to cecha porządkowa,
ponieważ możemy określić kolejność rozmiarów XL > L > M. 
\item Cechy nominalne nie definiują kolejności próbek. Przykładowo, można
za tego typu cechę uznać kolor koszulki, ponieważ zazwyczaj stwierdzenie, że czerwony
jest większy od niebieskiego, nie ma sensu.
\end{enumerate}

Poniższa tabela \textit{DataFrame}  składa się z cechy nominalnej (Kolor), porządkowej
(Rozmiar) i numerycznej (Cena). W ostatniej kolumnie są przechowywane etykiety klas (załóżmy, że
stworzyliśmy zestaw danych przeznaczonych do uczenia nadzorowanego). 

```python
df = pd.DataFrame([['Zielony', 'M', 10.1, 'klasa2'],
                   ['Czerwony', 'L', 13.5, 'klasa1'],
                   ['Niebieski', 'XL', 15.3, 'klasa2']])

df.columns = ['Kolor', 'Rozmiar', 'Cena', 'Etykieta klas']
df
```

<br>
<br>

## 2.2. Mapowanie cech porządkowych

Aby się upewnić, że algorytm uczenia poprawnie interpretuje cechy porządkowe, musimy przekształcić
wartości kategorialne występujące jako ciągi znaków w postać liczb całkowitych. Niestety,
nie istnieje żadna funkcja, która automatycznie przeniosłaby właściwy szyk etykiet z kolumny
Rozmiar, dlatego musimy własnoręcznie zdefiniować mapowanie cech. W poniższym przykładzie
zakładamy, że znamy różnice numeryczne pomiędzy cechami, np. XL = L+1 = M+2:

```python
size_mapping = {'XL': 3,
                'L': 2,
                'M': 1}

df['Rozmiar'] = df['Rozmiar'].map(size_mapping)
df
```

Przekształenie liczb całkowitych z powrotem na ich znakową reprezentację
(w którymś z późniejszych etapów) można wykonać  definiując słownik do odwrotnego mapowania

\textit{inv_size_mapping = \{v: k for k, v in size_mapping.items()\}}. 

Następnie  można go zastosować za pomocą metody $map$ do zmodyfikowanej kolumny cech:

```python
inv_size_mapping = {v: k for k, v in size_mapping.items()}
df['Rozmiar'].map(inv_size_mapping)
```

## 2.3. Kodowanie etykiet klas

Większość estymatorów klasyfikacji w interfejsie scikit-learn zawiera
wewnętrzne mechanizmy przekształcające etykiety klas w liczby całkowite, jednak zawsze
dobrym rozwiązaniem jest własnoręczne przygotowanie tych etykiet jako tablic z liczbami
całkowitymi w celu uniknięcia potencjalnych problemów technicznych. Do kodowania etykiet
klas można wykorzystać mechanizm podobny do
mapowania cech porządkowych. Należy pamiętać, że etykiety klas nie są cechami porządkowymi
i nie ma znaczenia, jaką liczbę całkowitą przydzielimy do danego ciągu znaków.
Możemy więc zwyczajnie ponumerować etykiety klas, począwszy od wartości 0:

```python
# tworzy słownik mapowania
# przekształcający etykiety klas z ciągów znaków do postaci liczb całkowitych
class_mapping = {label: idx for idx, label in enumerate(np.unique(df['Etykieta klas']))}
class_mapping
```

Teraz możemy wykorzystać słownik mapowania do przekształcenia etykiet klas w liczby całkowite:

```python
# przekształca etykiety klas z ciągów znaków do postaci liczb całkowitych
df['Etykieta klas'] = df['Etykieta klas'].map(class_mapping)
df
```

Możemy odwrócić parę klucz – wartość w słowniku mapowania, aby przywrócić etykietom
klas pierwotne wartości w postaci ciągów znaków:

```python
# odwraca proces mapowania
inv_class_mapping = {v: k for k, v in class_mapping.items()}
df['Etykieta klas'] = df['Etykieta klas'].map(inv_class_mapping)
df
```

Uzyskamy ten sam efekt, gdy skorzystamy z wygodnej klasy LabelEncoder zaimplementowanej
bezpośrednio w bibliotece scikit-learn:

```python
from sklearn.preprocessing import LabelEncoder

# kodowanie etykiet za pomocą klasy LabelEncoder
class_le = LabelEncoder()
y = class_le.fit_transform(df['Etykieta klas'].values)
y
```

Metoda $fit\_transform$ stanowi jedynie skrót do oddzielnego wywołania
metod $fit$ i $transform$, a metodę $inverse\_transform$ możemy wykorzystać do przekształcania
liczb całkowitych z powrotem do reprezentacji etykiet klas w postaci ciągów znaków:

```python
# odwrotne mapowanie
class_le.inverse_transform(y)
```

## 2.4. Kodowanie „gorącojedynkowe” cech nominalnych

Próbki znajdujące się w nominalnej kolumnie \textit{Kolor} trzeba przekształcić inaczej, gdyż zastowanie powyższych technik, tzn. 

```python
X = df[['Kolor', 'Rozmiar', 'Cena']].values
color_le = LabelEncoder()
X[:, 0] = color_le.fit_transform(X[:, 0])#przekształcamy zerową kolumnę, czyli dane nominalne
X
```

spowoduje, że po uruchomieniu powyższego kodu pierwsza kolumna w tabeli NumPy przechowuje teraz nowe
wartości cechy Kolor, które zostają zakodowane w następujący sposób:
Niebieski = 1, Zielony = 2, Czerwony = 0.

Gdybyśmy w tym momencie się zatrzymali i przesłali tabelę do klasyfikatora, popełnilibyśmy
jeden z najczęstszych błędów spotykanych podczas przetwarzania danych kategoryzujących.Mimo że wartości kolorów nie pojawiają się w określonej kolejności, algorytm uczący będzie teraz zakładał, że zielony jest większy od niebieskiego,
a niebieski od czerwonego. To założenie jest niewłaściwe, a jednak algorytm może generować
poprawne wyniki, ale mogą być one  dalekie od optymalnych.

Popularnym rozwiązaniem tego problemu jest zastosowanie techniki zwanej kodowaniem
„gorącojedynkowym” (ang. one-hot encoding). Koncepcją kryjącą się za tą metodą jest wprowadzenie
sztucznej cechy (ang. dummy feature) dla każdej unikatowej wartości w kolumnie cechy
nominalnej. W naszym przypadku możemy przekonwertować cechę Kolor w trzy nowe cechy:
Niebieski, Zielony i Czerwony. Wykorzystamy wartości binarne do wskazywania danego koloru
przykładu; np. niebieski przykład może zostać zakodowany jako Niebieski=1, Zielony=0,
Czerwony=0. Aby dokonać takiej transformacji, możemy wykorzystać klasę $OneHotEncoder$ zaimplementowaną
w module scikit-learn.preprocessing:

```python
from sklearn.preprocessing import OneHotEncoder

X = df[['Kolor', 'Rozmiar', 'Cena']].values
color_ohe = OneHotEncoder()
color_ohe.fit_transform(X[:, 0].reshape(-1, 1)).toarray()
```

```python
from sklearn.compose import ColumnTransformer

X = df[['Kolor', 'Rozmiar', 'Cena']].values
c_transf = ColumnTransformer([ ('onehot', OneHotEncoder(), [0]),
                               ('nothing', 'passthrough', [1, 2])])
c_transf.fit_transform(X).astype(float)
```

```python
# kodowanie gorącojedynkowe za pomocą biblioteki pandas

pd.get_dummies(df[['Kolor', 'Rozmiar', 'Cena']])
```

```python
# ochrona współliniowości w funkcji get_dummies

pd.get_dummies(df[['Kolor', 'Rozmiar', 'Cena']], drop_first=True)
```

```python
# ochrona współiniowości dla klasy OneHotEncoder

color_ohe = OneHotEncoder(categories='auto', drop='first')
c_transf = ColumnTransformer([ ('onehot', color_ohe, [0]),
                               ('nothing', 'passthrough', [1, 2])])
c_transf.fit_transform(X).astype(float)
```

# 3. Wielowarstwowe jednokierunkowe sieci neuronowe

Wielowarstwowa jednokierunkowa sieć neuronowa (ang. multi-layer feedforward neural network), często nazywana  również wielowarstwowym perceptronem (ang. multi-layer perceptron — MLP) naturalnie uogólnia koncepcję perceptronu. Istotną różnicą jest stosowanie różniczkowalnych funkcji aktywacji.

Poniższy rysunek przedstawia koncepcję trójwarstwowego modelu MLP; zawiera on warstwę wejściową,
warstwę ukrytą oraz warstwę wyjściową. Jednostki warstwy ukrytej są w pełni połączone
z jednostkami warstwy wejściowej, a jednostki warstwy wyjściowej — z jednostkami warstwy
ukrytej. Jeżeli taki model ma więcej niż jedną warstwę ukrytą, nazywamy go głęboką siecią
neuronową.

W ogólności będziemy oznaczać i-tą jednostkę aktywacji w l-tej warstwie jako $a_i^{(l)}$, a całą warstwę jednostek jako wektor $a^{(l)}$. Jednostka zerowa w warstwie to jednostka obciążenia (bias), przy czym warstwa wyjściowa nie zawiera takiej jednostki. Na poniższym poglądowym rysunku upraszczamy jednak oznaczenia.  Warstwę wejściową oznaczamy indeksem górnym $we$, warstwę ukrytą indeksem $u$, a warstwę wyjściową — indeksem $wy$. Dla warstwy wejściowej oczywiście zachodzi $a^{(we)}=x^{(we)}$.

Należy zwrócić uwagę, że jednostkę obciążenia (bias) traktujemy w sieciach niezależnie i zwykle mówiąc o warstwie sieci (na pozimie jej budowy bądź używania) nie wliczamy jej do ilości neurownów w warstwie (dlatego w schematach ma ona indeks zero).

```python
Image(filename='rysunki/12_02.png', width=600) 
```

## 3.1 Aktywacja sieci neuronowej za pomocą propagacji w przód


W ogólności uczenie modelu  sieci MLP można podzielić modelu na trzy podstawowe etapy:

\begin{enumerate}
	\item Począwszy od warstwy wejściowej, rozsyłamy do przodu wzorce danych uczących
wzdłuż całej sieci w celu uzyskania odpowiedzi.
\item Na podstawie otrzymanego wyniku obliczamy błąd, który chcemy
zminimalizować za pomocą funkcji kosztu opisanej w dalszej części rozdziału.
\itemRozprzestrzeniamy wstecz otrzymany błąd, wyliczamy jego pochodną przy
uwzględnieniu wszystkich wag, a następnie aktualizujemy model.
\end{enumerate}

Po powtórzeniu tych czynności dla wielu epok oraz wytrenowaniu wag modelu MLP,
można wykorzystać propagację w przód w celu obliczenia odpowiedzi, a także wprowadzić
funkcję progową po to, aby otrzymać przewidywane etykiety klas w notacji „gorącojedynkowej”,
omówionej w poprzednim podrozdziale.

W praktyce obliczenia 'w przód' w sieci MLP są wykonywane według następującego schematu.
\begin{enumerate}
	\item \textbf{Kalkulator warstwy wejściowej} - załóżmy, że w warstwie wejściowej jest $N_{in}$ neuronów
		\begin{enumerate}
			\item każdy z $N_{in}$ neuronów pobiera jedną  liczbę  $x_i (i=1,\dots,N_{in})$ oznaczającą dane wejściowe
			\item neurony niczego nie obliczają
			\item każdy z $N_{in}$ neuronów zwraca jedną  liczbę  oznaczającą dane wejściowe
		\end{enumerate}
	\item \textbf{Kalkulator międzywarstwowy} (pomiędzy dowolnymi dwoma warstwami, niezależnie od ich typu) 
			\begin{enumerate} 
				\item zakładamy, że warstwa wejściowa dla kalkulatora ma $N_h$ neuronów, a warstwa wyjściowa ma $M_h$ neuronów
				\item  pobiera wektor $N$  liczb $x = (x_i)_{i=1}^{N_h}$  zwróconych przez warstwę wejściową
				\item  pobiera macierz wag $W=w_{ij}$, $i=1,\dots,{N_h}$, $j=1,\dots,M_h$
				\item  pobiera wektor obciążeń $b = (b_j)_{j=1}^{M_h}$ neuronów 
				\item  oblicza wektor wejściowy  dla warstwy wyjściowej $y = \sigma(y_i)_{j=1}^{M_h}= W\cdot x + b$
			\end{enumerate}				
	\item \textbf{Warstwa ukryta i wyjściowa} - zakładamy, że warstwa zawiera $M_{out}$ neuronów
			 \begin{enumerate}
					\item  każdy z $M_{out}$ neuronów pobiera $N_{out}$  liczb $(y_i)_{i=1}^{N_{out}}$  oznaczających dane wejściowe dla tej warstwy
					\item  pobiera funkcję aktywacji $\sigma$ oraz  dla każdego $j=1,\dots,M_{out}$ oblicza $\sigma(y_j)$ 
					\item  zwraca wektor $(\sigma(y_j))_{j=1}^{M_{out}}$
			 \end{enumerate}	
	\item \textbf{Warstwa wyjściowa} 
				\begin{enumerate} 
					\item  dodatkowo oblicza wydajność sieci przez funkcję kosztu, której postać musi być dopasowana do funkcji aktywacji.
        \end{enumerate}
\end{enumerate}

```python
Image(filename='rysunki/12_03.png', width=500) 
```

Każda jednostka warstwy ukrytej jest powiązana ze wszystkimi jednostkami warstwy wejściowej, dlatego możemy obliczyć jednostkę
aktywacji $a_1^{(u)}$ warstwy ukrytej w następujący sposób:
$$
z_1^{(u)}=\sum\limits_{i=0}^m a_i^{(we)}w^{(u)}_{i,1}
$$
$$
a_1^{(u)}=\Phi(z_i^{(u)})
$$
przy czym $z_i^{(u)}$ jest pobudzeniem całkowitym, a $\Phi$ — funkcją aktywacji, która musi być różniczkowalna
w celu uczenia metodą gradientową wag łączących neurony. Przykładem takiej funkcji jest funkcja logistyczną (sigmoidalną), przyjmująca kształt krzywej s-kształtnej odwzorowującej
przekształcenie całkowitego pobudzenia $z$ w rozkład logistyczny mieszczący się w przedziale od
0 do 1; krzywa ta przechodzi przez oś y w punkcie z = 0 - zob. poniższy rysunek.

```python
Image(filename='rysunki/12_04.png', width=500) 
```

Aktywację całej warstwy zapisujemy macierzowo jako
$$
z^{(u)}=a^{(we)}W^{(u)} 
$$
$$
a^{(u)}=\Phi(z^{(u)})
$$
gdzie $a^{(we)}$ oznacza 1xm-wymiarowy wektor cech dla próby $x^{(we)}$ wraz z jednostką obciążenia (cechy zapisujemy w kolumnach),
$W^{(u)}$ jest mxd-wymiarową macierzą wag, gdzie \textit{ d} definiuje liczbę jednostek w warstwie ukrytej. Po
przeprowadzeniu mnożenia macierzowo-wektorowego otrzymujemy 1xd-wymiarowy wektor
całkowitego pobudzenia $z^{(u)}$ umożliwiający obliczenie aktywacji $a^{(u)}$, gdzie $a^{(u)}$ jest 1xd-wymiarowym wektorem, tzn. aktywację wykonujemy oddzielnie dla każdej składowej  wektora $z^{(u)}$ (symbolicznie zapisujemy to jako $\Phi(z^{(u)})$) 

W powyższym zapisie pobieramy jedną próbkę jako wektor wierszowy i obliczamy jej obraz w jednej warstwie sieci, otrzymując jako wynik wektor wierszowy. Przyjmijmy teraz, że mamy w ogólności \textit{n} próbek, które upakowujemy w macierz $A^{(we)}$ wymiaru
nxm. Możemy wówczas obliczyć odpowiedź pojedynczej warstwy dla wszystkich próbek jednocześnie, stosując mnożenie macierzy
$$
Z^{(u)}=A^{(we)}W^{(u)} 
$$
Macierz wynikowa $Z^{(u)}$ jest wymiaru nxd, tzn. każdy jej wiersz odpowiada wynikowi pobudzenia jednej próbki. Podobnie jak powyżej stosujemy symboliczny zapis aktywacji warstwy na wszystkich próbkach:
$$
A^{(u)}=\Phi(Z^{(u)})
$$
otrzymując nxd wymiarową macierz aktywacji $A^{(u)}$, w której pojedynczy wiersz odpowiada aktywacji pojedynczej próbki przez warstwę.

Analogiczną strategię stosujemy do każdej pary kolejnych dwóch warstw w wielowarstwowej sieci jednokierunkowej.

## 3.2. Zestaw danych pisma odręcznego z bazy MNIST

Zestaw danych MNIST jest dostępny publicznie pod adresem http://yann.lecun.com/exdb/mnist/
i składa się z czterech części:

\begin{enumerate}

Zestaw danych MNIST został utworzony z dwóch zbiorów danych amerykańskiego Narodowego

szkoły średniej, a druga połowa przez pracowników amerykańskiego Biura Cenzusowego.
Taka sama proporcja dotyczy próbek tworzących zestaw danych testowych.

Obrazy są przechowywane w formacie bajtowym i będziemy je wczytywać do tablic NumPy, które
następnie posłużą nam do uczenia i testowania implementacji MLP. W tym celu zaimplementujemy
następującą funkcję pomocniczą:

```python
import os
import struct
import numpy as np
 
def load_mnist(path, kind='train'):
    """Wczytuje dane MNIST umieszczone w katalogu roboczym"""
    labels_path = os.path.join(path, 
                               '%s-labels.idx1-ubyte' % kind)
    images_path = os.path.join(path, 
                               '%s-images.idx3-ubyte' % kind)
        
    with open(labels_path, 'rb') as lbpath:
        magic, n = struct.unpack('>II', 
                                 lbpath.read(8))
        print(n)
        labels = np.fromfile(lbpath, 
                             dtype=np.uint8)

    with open(images_path, 'rb') as imgpath:
        magic, num, rows, cols = struct.unpack(">IIII", 
                                               imgpath.read(16))
        images = np.fromfile(imgpath, 
                             dtype=np.uint8).reshape(len(labels), 784)
        images = ((images / 255.) - .5) * 2
 
    return images, labels
```

Funkcja \textit{load\_mnist}  zwraca dwie tablice:
\begin{enumerate}
\item Pierwsza (\textit{images}) z nich to n×m-wymiarowa tablica NumPy, gdzie n oznacza liczbę przykładów, a m — cech (w tym przypadku pikseli). Zestaw
danych uczących składa się z 60 000 przykładów, a danych testowych z 10 000 cyfr.
Poszczególne rysunki mają rozmiar 28×28 pikseli; każdemu pikselowi przyporządkowana
jest wartość odcienia szarości. W naszym przypadku rozwijamy te rysunki na jednowymiarowe
wektory symbolizujące rzędy w macierzy images (784 punkty na każdy rząd lub rysunek).
\item Druga tabela (\textit{labels} ) zwracana przez funkcję \textit{load_mnist}  zawiera odpowiednie zmienne docelowe
— etykiety klas (liczby całkowite od 0 do 9) dla rysunków odręcznie pisanych cyfr.
\end{enumerate}

## 3.3. Algorytmy uczenia


Błąd modelu (czyli funkcja kosztu) powinien  być dobrym estymatorem jakości modelu budowanego na ustalonych  danych. W sekcji tej wyjaśnimy dokładnie tylko jedną koncepcję uczenia maszynowego, gdyż jest ona stosowana w większości sieci neuronowych.  Szczegóły pozostałych podejść do tego zagadnienia takie jak SVM, knn, drzewa decyzyjne, PCA oraz metoda k-średnich można znaleźć w klasycznej literaturze dotyczącej uczenia maszynowego.

Załóżmy, że rozpatrujemy zbiór danych wygenerowanych niezależnie $(x_1,x_2,\dots,x_m)$. Naturalne jest założenie, że rozkład prawdopodobieństwa $p_{data}(x)$ tych danych  jest jednakowy, tzn. dla każdego $i\not = j$ $p_{data}(x_i)=p_{data}(x_j)$. Mówimy wtedy, że dane są $i.i.d.$ (independent identically distributed).  Załóżmy, że celem jest estymacja parametru $\theta$ (w sieciach neuronowych $\theta$ jest wektorem wszystkich wag). Teoretycznym estymatorem tego parametru jest każda funkcja $\hat{\theta}_m = g(x_1,x_2,\dots,x_m)$. Jest to zmienna losowa, gdyż dane traktujemy jako wylosowane. Rozkład $p_{data}$ jest oczywiście nieznany, gdyż inaczej budowanie modelu byłoby bezcelowe. Zwykle jednak zakładamy, że rozkład danych należy do pewnej klasy rozkładów o nieznanych parametrach $\theta$ rozkładu. Przykładowo dla danej $x$, o której zakładamy, że jest próbką z rozkładu normalnego parametr $\theta$ jest dwuwymiarowy:
$\theta=(\mu,\sigma)$. 
Model możemy zatem traktować jako estymator nieznanego rozkładu danych  opracowany na próbce takich danych (skończonym podzbiorze, który znamy). Oznaczmy przez $p_{model}(x;\theta)$ prawdopodobieństwo rozkładu estymującego za pomocą modelu  rzeczywisty nieznany rozkład $p_{data}(x)$ dla ustalonej próbki $x$. Im takie prawdopodobieństwo  rozkładu modelu dla ustalonej próbki będzie wyższe, tym lepiej model będzie reprezentował tą próbkę. Rozpatrując wszystkie $m$ próbek jednocześnie  musimy rozpatrywać ich prawdopodobieństwo wspólne, tzn. $p_{model}(x_1,x_2,\dots,x_m; \theta)$. Z niezależności próbek dostajemy, że 

$$p_{model}(x_1,x_2,\dots,x_m; \theta)= p_{model}(x_1;\theta)\cdot \dots p_{model}(x_m;\theta) = \prod_{i=1}^m p_{model}(x_i; \theta)$$. \\

\noindent Te rozważania prowadzą do zdefiniowania funkcji nazywanej \textit{ funkcją wiarygodności estymatora z parametrem $\theta$}:
$$l(\theta) = K\cdot\prod_{i=1}^m p_{model}(x_i;\theta),$$
gdzie $K$ jest pewną stałą -  współczynnikiem proporcjonalności. \textit{Estymatorem największej wiarygodności parametru} $\theta$ przy ustalonych danych $x_1,x_2,\dots,x_m$ nazywamy taką
funkcję wiarygodności $l(\hat{\theta})$,  że
$$
l(\hat{\theta})= \max_{\theta}l(\theta).
$$

Równoważnie

$$
\hat{\theta} =arg\ \max_{\theta}l(\theta) = arg\ \max_{\theta} \prod_{i=1}^m p_{model}(x_i;\theta).
$$


W związku z tym, że każdy czynnik w iloczynie prawdopodobieństw jest nieujemnym ułamkiem mniejszym od 1, numeryka takiego iloczynu implikuje zagrożenie niedomiarem 
zmiennoprzecinkowym.  Możemy to zagrożenie usunąć korzystając z prostego faktu:

##### Fakt pomocniczy:
Niech $0>a>b$.  Niech $f:[a,b]\rightarrow\mathbb{R}$, $g:\mathbb{R}\rightarrow\mathbb{R}$ będą ciągłe na $[a,b]$ oraz $g(x)=ln(f(x))$ dla $x\in \mathbb{R}$. Wtedy funkcje $f$ oraz $g$ osiągają wartości ekstremalne w tych  w tych 
samych punktach.


\textbf{Uzasadnienie}: Z ciągłości i zwartości przedziału $[a,b]$ dostajemy, że $f$ i $g$ osiągają ekstrema na $[a,b]$. $g'(x) = (ln(f(x)))' = \frac{f'(x)}{f(x)}$. 
Wtedy jeżeli $g'(x)$ istnieje oraz  $g'(x)=0$, to $f'(x)=0$, co kończy dowód.    

Tak sformułowany Fakt pomocniczy pozwala zadanie maksymalizacji funkcji wiarygodności zastąpić  zadaniem maksymalizacji logarytmu tej funkcji:
\begin{eqnarray}
\hat{\theta} &=& arg\ \max_{\theta} \prod_{i=1}^m p_{model}(x_i;\theta) =arg\  \max_{\theta} log\,\left(\prod_{i=1}^m p_{model}(x_i;\theta).\right) \nonumber \\
&=& arg\ \max_{\theta} \sum_{i=1}^m log(p_{model}(x_i;\theta)\label{eq:ml2}
\end{eqnarray}
  
Logarytmiczna formuła estymatora największej wiarygodności nie jest już niebezpieczna ze względu na niedomiar zmiennoprzecinkowy. Definiując rozkład empiryczny zbioru próbek
jako
$$
\hat{p}_{data}(x) = \frac{1}{m}\sum_{i=1}^m x_i
$$
i dzieląc prawą stronę równania (13) przez $m$ możemy (stała nie zmienia szukania maksimum)  równanie (13) zapisać jako
$$
\hat{\theta} = arg \ \max_{\theta} E_{x \sim \hat{p}_{data}} log(p_{model}(x;\theta)
$$

Widać tu lepiej, że istotą największej wiarygodności jest doprowadzenie rozkładu modelu możliwie blisko rozkładowi  empirycznemu danych szkoleniowych. Jest to oczywiście słabsze żądanie niż 
optymalne zbliżenie rozkładu modelu do rzeczywistego rozkładu $p_{data}(x)$, ale niczego poza próbką danych nie posiadamy. 

Tak zdefiniowane zadanie można przeformułować do postaci jawnie pokazującej rozbieżność informacyjną pomiędzy rozkładem empirycznym danych szkoleniowych  i rozkładem modelu. Służy do tego tzw. \textit{ dywergencja Kullbacka-Leiblera}, w tym przypadku mająca postać:
$$
d_{KL}(p_{model}, \hat{p}_{data}) = E_{x \sim \hat{p}_{data}} \left(log(\hat{p}_{data}(x) - log(p_{model}(x;\theta)\right)
$$
Zadanie (15) jest w tym przypadku równoważne minimalizacji dywergencji Kullbacka-Leiblera według parametru $\theta$.  Równanie (16) po rozpisaniu przyjmuje postać
$$
d_{KL}(p_{model}, \hat{p}_{data}) = \sum_{i=1}^m \hat{p}_{data}(x_i)\,log(\hat{p}_{data}(x_i) - \sum_{i=1}^m \hat{p}_{data}(x)\,log(p_{model}(x_i;\theta),
$$
zatem minimalizacja dywergencji Kullbacka-Leiblera według parametru $\theta$ oznacza minimalizację drugiego składnika prawej strony równania (17). Ostatecznie, maksymalizacja wiarygodności oznacza minimalizację wielkości 
$$
-E_{x \sim \hat{p}_{data}} log(p_{model}(x;\theta).
$$


Pierwszy składnik prawej strony równania (17) to definicyjnie ujemna entropia informacyjna $H(\hat(p)_{data})$. Zatem równanie (17) można zapisać jako
$$
-E_{x \sim \hat{p}_{data}} log(p_{model}(x;\theta)) = H(\hat{p}_{data}) + d_{KL}(p_{model}, \hat{p}_{data}).
$$
Wyrażenie z lewej strony byłoby równe entropii $H(\hat{p}_{data})$, gdyby rozkład $p_{model}$ był równy rozkładowi $\hat{p}_{data}$. Zatem dywergencję Kullbacka-Leiblera  można interpretować jako miarę zaburzenia informacji, którą wnosi zastosowanie niedokładnego rozkładu  $p_{model})$ w miejsce do rzeczywistego   rozkładu  $\hat{p}_{data}$. Wyjaśnia to dlaczego wyrażenie 
$-E_{x \sim \hat{p}_{data}} log(p_{model}(x;\theta))$ nazywane jest \textit{entropią krzyżową} (czyli w istocie 'informacją krzyżową'). \\

\noindent \textbf{Konkluzja}: Zadanie maksymalizacji wiarygodności jest równoważne zadaniu minimalizacji entropii krzyżowej.
 

Entropia krzyżowa jest naturalnym źródłem konstrukcji różnych funkcji straty:
\begin{enumerate}
\item Wybieramy jako klasę modelu zbiór rozkładów Gaussa, tzn. $\theta = (\mu,\sigma)$, gdzie $\mu$ jest wartością oczekiwaną rozkładu, a $\sigma$ jego wariancją. Rozpatrujemy zbiór przykładów
uczących $(x,y) = (x_i,y_i)_{i=1}^n$ oraz zakładamy, że wariancja $\sigma$ danych szkoleniowych jest stała i że model aproksymuje wartość oczekiwaną rozkładu empirycznego danych szkoleniowych, 
tzn. $\mu\approx \hat{y}(x,w)$. Wtedy 

$$
p_{model}(y|x)= N(y;\hat{y}(x,w),\sigma^2) = \frac{1}{2\pi\sigma^2}e^{\frac{\|y_i-\hat{y_i}(x_i,w)\|^2}{2\sigma^2}}.
$$  
Zatem 

$$
-E_{x,y \sim \hat{p}_{data}} log(p_{model}(x;\theta))  = -\frac{1}{m}\sum_{i=1}^m log\left(\frac{1}{\sqrt{2\pi}\sigma} e^{-\frac{\|y_i-\hat{y_i}(x_i,w)\|^2}{2\sigma^2}}\right)   
$$

$$
=-\frac{1}{m}m\,log\left(\frac{1}{2\pi\sigma^2}\right) -\frac{1}{m}\sum_{i=1}^m log\left(e^{-\frac{\|y_i-\hat{y_i}(x_i,w)\|^2}{2\sigma^2}}\right) 
$$
$$
=-log(1) + \frac{1}{2}log(2\pi) + log(\sigma) -\frac{1}{m}\sum_{i=1}^m log\left(e^{-\frac{\|y_i-\hat{y_i}(x_i,w)\|^2}{2\sigma^2}}\right) 
$$


$$
=\frac{1}{2}log(2\pi) + log(\sigma) +\frac{1}{m}\sum_{i=1}^m \frac{\|y_i-\hat{y_i}(x_i,w)\|^2}{2\sigma^2} 
$$
$$
=\frac{1}{2}log(2\pi) + log(\sigma) +\frac{1}{2m\sigma^2}\sum_{i=1}^m \|y_i-\hat{y_i}(x_i,w)\|^2
$$
Widać zatem, że minimalizacja takiej wielkości względem parametrów modelu, którymi w tym przypadku są wagi $w$ odpowiada minimalizacji ostatniego składnika. Wynika z tego, że minimalizacja 
entropii krzyżowej pomiędzy rozkładem empirycznym danych szkoleniowych i rozkładem Gaussa  jest równoważna minimalizacji błędu średniokwadratowego. W szczególności oznacza to , że regresja liniowa 
jest metodą maksymalizującą wiarygodność. Dodatkowym wnioskiem jest tu uwaga, że sieć aproksymująca średnią z danych szkoleniowych (wartość oczekiwana jest przecież średnią) nie wymaga dla zadania minimalizacji funkcji kosztu żadnych metod optymalizacji (typu metody gradientowe), gdyż jak wiadomo minimum jest obliczalne przez zwykły układ równań. 

Zwróćmy w tym miejscu jeszcze uwagę na ciekawe cechy estymatora średniokwadratowego (ozn. MSE), a mianowicie na jego zgodność i obciążoność. 

$$	
MSE(\theta,\theta_m) = E((\theta-\theta_m)^2) = E(\theta^2) + E(\theta_m^2) - 2*E(\theta*\theta_m)
$$

$$
Bias(\theta_m) = E(\theta) - E(\theta_m)  
$$

$$
(Bias(\theta_m))^2 = (E(\theta) - E(\theta_m))^2 = (E(\theta))^2 +  (E(\theta_m))^2 - 2* E(\theta)*E(\theta_m)
$$
	
$$
Var(\theta_m) = E((\theta_m-E(\theta_m))^2)= E(\theta_m^2) + (E(\theta_m))^2 - 2*(E(\theta_m))^2 =
$$

$$
= E(\theta_m^2) - (E(\theta_m))^2
$$

Stąd 
	
$$
(Bias(\theta_m))^2 + Var(\theta_m) =
$$

$$
= (E(\theta))^2 +  (E(\theta_m))^2 - 2* E(\theta)*E(\theta_m) + E(\theta_m^2) - (E(\theta_m))^2 =
$$

$$  
= (E(\theta))^2  - 2* E(\theta)*E(\theta_m) + E(\theta_m^2) = MSE(\theta,\theta_m)
$$
	 
	 
Zatem $MSE(\theta,\theta_m)$ jest estymatorem jakości modelu $\theta_m$ sumującym w sobie obciążenie i wariancję $\theta_m$. 


Obciążoność modelu pokazuje, jak daleko, średnio, predykcja estymatora odchyla się od rzeczywistej średniej.
Wariancja modelu odzwierciedla wrażliwość predykcji na zbiór uczący. Zarówno obciążoność, jak i wariancja  zależą rownież od konkretnego rozkładu, do którego stosowany jest model.

W wielu modelach regresji występuje parametr kontrolujący ich elastyczność
(lub złożoność). Na przykład elastyczność sieci neuronowych zależy od
liczba ukrytych warstw i liczby znajdujących się w nich neuronów. Elastyczność
klasyfikatora k-najbliższych sąsiadów zależy od liczby k. Im bardziej elastyczny
model, tym większa jest przestrzeń możliwych do rozpatrzenia funkcji . Takie modele
mają mniejszą obciążoność, ale większą wariancję. Stąd pojawia się problem ze znalezieniem
optymalnej wartości elastycznosci, która daje optymalną równowagę między obciążonością i
wariancją. 

Poniższy rysunek pokazuje reprezentacje czterech  różnych scenariuszy. Okręgi
reprezentują poziomy funkcji straty. Mniejsze kółka oznaczają mniejszą jej wartość. Każdy zbiór  kropek odpowiada innemu zestawowi treningowemu. Wysoka obciążoność oznacza duży błąd, wysoka wariancja nie gwarantuje niskiego błędu.

![bias_variance](.\nn_images\bias_variance_reg.png)

\item Rozpatrzmy model z klasy rozkładów Bernoulli'ego, czyli dwuelementowego zbioru wartości, konwencjonalnie 0 lub 1. Puentą przykładu wstępnego i późniejszych uwag o potrzebie wygładzania funkcji aktywacji dla alogrytmów gradientowych było wprowadzenie w takiej sytuacji funkcji logistycznej (ozn. $\sigma$) jako funkcji aktywacji. W przypadku klasyfikacji binarnej z wartościami 0,1, oprócz gładkości potrzebujemy również zakresu wartości [0,1], co jest zapewnione w konstrukcji funkcji logistycznej. Ponadto można ją zdefiniować na dowolnej dziedzinie, więc w szczególności takiej, na której całka będzie równa 1, co pozwala taką funkcję traktować jako gęstość prawdopodobieństwa (w klasyfikacji binarnej wartość 0.5 jest używana jako próg klasyfikujący). W efekcie 
$\hat{y}(x)= \sigma(w^Tx+b)$ oraz

 $$
-E_{x,y \sim \hat{p}_{data}} log(p_{model}(x;\theta))  = -\frac{1}{m}\sum_{i=1}^m y_i log(\hat{y_i})+  (1- y_i)log(1-\hat{y_i}).
$$
Tak skonstruowaną funkcję starty nazywamy regresją logistyczną\footnote{Niektórzy autorzy zwracają uwagę na dziwność tej nazwy, gdyż stosuje się ją do zagadnienia klasyfikacyjnego} i 
stosujemy ją w do klasyfikacji binarnej w parze z funkcją logistyczną jako funkcją aktywacji.
\item W przypadku klasyfikacji wielowartościowej uogólniamy procedurę z poprzedniego punktu. Rozpatrujemy
cały wektor wartości $(y_i)_{i=1}^m$  i  definiujemy na nim  funkcję aktywacji  (tzw. \textit{softmax}) zwracającą wektor takiego samego rozmiaru, w który i-ta składowa ma postać:
$$
p(y_i) = \frac{e^{x_i}}{\sum_{k=1}^me^{x_k}}.
$$
Wektor taki może być traktowany jako wektor prawdopodobieństwa, co implikuje, że funkcja straty wyprowadzona z minimalizacji entropii krzyżowej jest identyczna jak w przypadku klasyfikacji binarnej.
\end{enumerate}

\textbf{Uwaga}: W sieciach stosujących ReLU jako funkcję aktywacji zwykle ostatnia warstwa stosuje funkcję aktywacji inną niż ReLU, najczęściej logistyczną lub softmax, co oznacza, że
można także w tym przypadku korzystać z entropii krzyżowej jako funkcji straty.


## 3.4. Regularyzacja w sieciach neuronowych

Luźno wyjaśnione w punkcie 3.3 pojęcie <i>elastyczności</i> modelu (mówimy też zamiennie o jego <i>złożoności</i>) może być sformułowane precyzyjnie. Jedną z możliwości jest zastosowanie <b>wymiaru Wapnika-Czerwonienkisa</b> (ozn. VC). 

Wiadomo już, że perceptron może poprawnie rozdzielić każde trzy niewspółliniowe punkty leżące na płaszczyźnie, tzn. przydzielić je jednoznacznie do dodaniej i ujemnej półpłaszczyzny pewnej prostej. Niemniej  ale w przypadku  czterech punktów już się tego nie da zrobić. Wynika to bezpośrednio z [Twierdzenia Radona](https://en.wikipedia.org/wiki/Radon%27s_theorem) Da się z tego wywnioskować, że VC perceptronu działającego na płaszczyźnie wynosi 3 ( wynosi <i>d+1</i> jeżeli perceptron działa w <i>d</i>-wymiarowej przestrzeni).

Bardziej formalnie VC jest zdefiniowany jako liczność największego zbioru, który jest <i>rozbijany</i> przez przestrzeń hipotez - w przypadku perceptronu przestrzeń hipotez stanowi zbiór hiperpłaszczyzn w danej przestrzeni. Takie rozbijanie oznacza, że istnieje hipoteza, która na zbiorze treningowym daje zerową wartość funkcji straty. Liczność zbioru jest oczywiście naturalnym parametrem dla myślenia o jego złożoności. 

VC umożliwia szacowanie błędu generowanego przez funkcję straty na zbiorze testowym (przy ustalonej hipotezie), korzystając z wartości błędu na zbiorze treningowym i pewnej formuły zawierającej wymiar VC. Oznaczając przez $f(x,w)$ wybraną hipotezę w sieci neuronowej formuła szacunkowa na błąd jej działnia ma postać:
$$
P\left(L_{test}(f)\leq L_{train}(f)+\sqrt{\frac{Hlog(2N/H)+H-log(\theta/4)}{N}} \right)) < 1-  \theta,
$$
gdzie $\theta\in(0,1) $, $L_{test}(f)$ oraz $L_{train}(f)$ oznaczają wartości funkcji straty na zbiorach, odpowiednio, testowym i treningowym przy ustalonej hipotezie $f$, $H$ jest wartością wymiaru VC, natomiast $N$ oznacza liczność zbioru treningowego. W ogólności VC zależy od wielu hiperparamterów sieci (np. liczy warstw i neuronów). 

Wysoka złożoność modelu (wyrażona w tym przypadku przez VC, czyli m.in. liczność zbioru treningowego) może spowodować <i>przeuczenie</i> (ang. overfitting) sieci, kiedy to model zbyt ściśle trzyma się z małym błędem zbioru treningowego.Powoduje to jego przesztywnienie i błąd na zbiorze testowym zaczyna rosnąć. Pokazuje to poniższy rysunek:

![overfit](nn_images/overfit.png)

Optymalna złożoność minimalizuje błąd testowy, przy czym, jak widać, oznacza to unikanie przeuczenia modelu. Dobrym narzędziem jest w tym przypadku tzw. <i>regularyzacja</i> modelu. Od razu widać, że jakakolwiek metoda regularyzacji będzie oznaczała unikanie wysokiej wariancji. Czasami można to zrobić niemal ręcznie. bezpośrednio manipulując postacią zbioru danych. Można też (i jest to częste) wykorzystać wiedzę dodatkową apriori (tzw. domain knowledge) o istotnych cechach szukanego rozwiązania (optymalnej hipotezy $f$). Wówczas do klasycznego zadania opytmalizacyjnego ($L(f(x;w))$ oznacza funkcję  straty)

$$
w^\star = arg\ min_w L_{train}(f(x;w)) 
$$

dodajemy jakiś dodatkowy warunek na szukane $f(x;w)$. Zwykle ta dodatkowa wiedza daje się wyrazić w postaci warunku liczbowego  i zwykle jest to jakiś rodzaj ograniczenia wartości tego warunku. Prowadzi to do sformułowania zadania w postaci problemu optymalizacji z ograniczeniami nierównościowymi:
$$
\left\{ \begin{array}{l}
w^\star = arg\ min_w L_{train}(f(x;w))\\
R(f)\leq C
\end{array}
\right.
$$

Przykładowo, w wielu przypadkach regresji wiemy, że $f(x;w)$  powinna być gładka. Formalnie oznacza to, że jego „gładkość” powinna być mniejsza niż konkretna wartość, co można wyrazić przez zsumowanie wartości odpowednich pochodnych we wszystkich punktach:
$$
R(f) = \int\left|\frac{\partial^m}{\partial x^m}f(x;w)\right|^2dx.
$$
Dla $m = 1$ ogranicza to $f(x;w)$ od  posiadania zbyt dużych pierwszych pochodnych, podczas gdy dla $m =
2$ ogranicza $f(x;w)$ przed zbyt szybką zmianą pierwszej pochodnej. Parametr
$C$ jest hiperparametrem regulującym siłę regularyzacji. Można go wybrać bardziej lub mniej precyzyjnie (por. szybkość uczenia w metodzie najszybszego spadku).

W przypadku sieci neuronowych, kiedy neuronów są tysiące lub miliony, naturalne jest podejrzenie, że część z tych neuronów jest mało istotna w finalnym modelu, tzn. ich finalne wagi są bardzo małe. Neurony te stanowią rodzaj szumu informacyjnego. Celowe byłoby zatem, aby je w ogóle wyeliminować, tzn. spowodować by ich wagi stały się zerowe.  Idealna jest wtedy jakaś norma 'globalna' na wektorze $\mathbf{w}$ przechowującym wszystkie wagi. Klasycznym wyborem są normy $L_1$ lub $L_2$, tzn.
$$
R(f(x;\mathbf{w})) = \|\mathbf{w}\|_1 \ \ \ \hbox{lub}\ \ \ \ R(f(x;\mathbf{w})) = \|\mathbf{w}\|_2.
$$

Wtedy (przy odpowiednim $C$) wagi docelowo małe znikają, a wagi istotne (duże) zachowują swoje istnienie. Z takiego punktu widzenia można na taką regularyzację spojrzeć jak na metodę redukcji wymiarowości zagadnienia (podobną w pomyśle do  np. PCA).

Z ogólnej teorii optymalizacji (Twierdzenie Kuhna-Tuckera) wiadomo, że problem optymalizacyjny (39) oznacza równoważnie minimalizację funkcji

$$
L_{train}(x;w)+\lambda R(f)
$$

Metoda największego spadku ma wtedy postać

$$
\mathbf{w}^k = \mathbf{w}^{k-1}+ \alpha\left(\frac{\partial L_{train}}{\partial \mathbf{w}}(f(\mathbf{w}))+\lambda \frac{\partial R}{\partial \mathbf{w}}(f(\mathbf{w}))\right) 
$$

Rysunki poniżej  (lewy dla normy $L_1$, prawy dla normy $L_2$) pokazują, że rzeczywiste minima wag przesuwają się po regularyzacji  bliżej zera - na zielono zaznaczono warstwice funkcji $L_{train}()$, na czerwono warstwice funkcji $R()$.

## 3.5. Implementacja perceptronu wielowarstwowego

Poniższy kod jest implementacją modelu MLP klasyfikującego obrazy z zestawu
danych MNIST. Dla uproszczenia zaimplementujemy model zawierający tylko jedną warstwę
ukrytą.


```python
class NeuralNetMLP(object):
    """ Sieć neuronowa z dodatnim sprzężeniem zwrotnym / klasyfikator – perceptron wielowarstwowy.

    Parametry
    ------------
    n_hidden : liczba całkowita (domyślnie: 30)
          Liczba jednostek ukrytych.
    l2 : wartość zmiennoprzecinkowa (domyślnie: 0.0)
          Parametr lambda regularyzacji L2.
          Nie ma regularyzacji, jeśli l2=0.0 (domyślnie)
    epochs : liczba całkowita (domyślnie: 500)
          Liczba przebiegów po zestawie danych uczących.
    eta : wartość zmiennoprzecinkowa (domyślnie: 0.001)
          Współczynnik uczenia.
    shuffle : typ boolowski (domyślnie: True)
          Jeżeli wartość jest równa "True", tasuje dane uczące po każdej epoce
          w celu uniknięcia cykliczności.
    minibatch_size : liczba całkowita (domyślnie: 1)
          Liczba przykładów uczących na daną mini-grupę.
    seed : liczba całkowita (domyślnie: None)
          Ziarno losowości używane podczas tasowania i inicjowania wag.

    Atrybuty
    -----------
    eval_: słownik
          Słownik przechowujący wartości kosztu, dokładności uczenia i dokładności
          walidacji dla każdej epoki podczas uczenia.

    """

    def __init__(self, n_hidden=30,
                 l2=0., epochs=100, eta=0.001,
                 shuffle=True, minibatch_size=1, seed=None):

        self.random = np.random.RandomState(seed)
        self.n_hidden = n_hidden
        self.l2 = l2
        self.epochs = epochs
        self.eta = eta
        self.shuffle = shuffle
        self.minibatch_size = minibatch_size

    def _onehot(self, y, n_classes):
        """Koduje etykiety do postaci "gorącojedynkowej"

        Parametry
        ------------
        y : tablica, wymiary = [n_próbek]
            Wartości docelowe.

        Zwraca
        -----------
        onehot : tablica, wymiary = (n_etykiet, n_próbek)

        """
        onehot = np.zeros((n_classes, y.shape[0]))
        for idx, val in enumerate(y):
            onehot[val, idx] = 1.
        return onehot.T

    def _sigmoid(self, z):
        """Oblicza funkcję logistyczną (sigmoidalną)"""
        return 1. / (1. + np.exp(-np.clip(z, -250, 250)))

    def _forward(self, X):
        """Oblicza krok propagacji w przód"""

        # Etap 1.: Pobudzenie całkowite warstwy ukrytej
        # [n_próbek, n_cech] dot [n_cech, n_hidden]
        # -> [n_próbek, n_hidden]
        z_h = np.dot(X, self.w_h) + self.b_h

        # Etap 2.: Aktywacja warstwy ukrytej
        a_h = self._sigmoid(z_h)

        # Etap 3.: Całkowite pobudzenie warstwy wyjściowej
        # [n_próbek, n_hidden] dot [n_hidden, n_etykiet_klas]
        # -> [n_próbek, n_etykiet_klas]

        z_out = np.dot(a_h, self.w_out) + self.b_out

        # Etap 4.: Aktywacja warstwy wyjściowej
        a_out = self._sigmoid(z_out)

        return z_h, a_h, z_out, a_out

    def _compute_cost(self, y_enc, output):
        """Oblicza funkcję kosztu.

        Parametry
        ----------
        y_enc : tablica,wymiary = (n_etykiet, n_próbek)
            Etykiety klas zakodowane do postaci "gorącojedynkowej".
        output : tablica, wymiary = [n_próbek, n_jednostek_wyjściowych]
            Aktywacja warstwy wyjściowej (propagacja w przód).

        Zwraca
        ---------
        cost : wartość zmiennoprzecinkowa
            Regularyzowana funkcja kosztu.

        """
        L2_term = (self.l2 *
                   (np.sum(self.w_h ** 2.) +
                    np.sum(self.w_out ** 2.)))

        term1 = -y_enc * (np.log(output))
        term2 = (1. - y_enc) * np.log(1. - output)
        cost = np.sum(term1 - term2) + L2_term

        # Jeżeli będziesz stosować tę funkcję kosztu do innych zestawów danych,
        # w których wartości aktywacji mogą przyjmować skrajniejsze
        # wartości (bliższe 0 lub 1), możesz natrafić na błąd
        # "ZeroDivisionError" z powodu niestabilności numerycznej
        # bieżących implementacji języka Python i biblioteki NumPy.
        # Przykładowo, algorytm stara się obliczyć log(0), czyli niezdefiniowaną wartość.
        # Aby rozwiązać ten problem, możesz dodać niewielką stałą
        # do wartości aktywacji; stała ta zostanie dodana do funkcji logarytmicznej.
        #
        # Na przykład:
        #
        # term1 = -y_enc * (np.log(output + 1e-5))
        # term2 = (1. - y_enc) * np.log(1. - output + 1e-5)

        return cost

    def predict(self, X):
        """Prognozowanie etykiet klas

        Parametry
        -----------
        X : tablica, wymiary = [n_próbek, n_cech]
            Warstwa wejściowa z pierwotnymi cechami.

        Zwraca:
        ----------
        y_pred : tablica, wymiary = [n_próbek]
            Przewidywane etykiety klas.

        """
        z_h, a_h, z_out, a_out = self._forward(X)
        y_pred = np.argmax(z_out, axis=1)
        return y_pred

    def fit(self, X_train, y_train, X_valid, y_valid):
        """ Aktualizuje wagi za pomocą danych uczących.

        Parametry
        -----------
        X_train : tablica, wymiary = [n_próbek, n_cech]
            Warstwa wejściowa zawierająca pierwotne cechy.
        y_train : tablica, wymiary = [n_próbek]
            Docelowe etykiety klas.
        X_valid : tablica, wymiary = [n_próbek, n_cech]
            Przykładowe cechy służące do walidacji w trakcie uczenia.
        y_valid : tablica, wymiary = [n_próbek]
            Przykładowe etykiety służące do walidacji w trakcie uczenia.

        Zwraca:
        ----------
        self

        """
        n_output = np.unique(y_train).shape[0]  # number of class labels
        n_features = X_train.shape[1]

        ########################
        # Inicjowanie wag
        ########################

        # wagi pomiędzy warstwą wejściową a ukrytą
        self.b_h = np.zeros(self.n_hidden)
        self.w_h = self.random.normal(loc=0.0, scale=0.1,
                                      size=(n_features, self.n_hidden))

        # wagi pomiędzy warstwą ukrytą a wyjściową
        self.b_out = np.zeros(n_output)
        self.w_out = self.random.normal(loc=0.0, scale=0.1,
                                        size=(self.n_hidden, n_output))

        epoch_strlen = len(str(self.epochs))  # formatowanie paska postępów
        self.eval_ = {'cost': [], 'train_acc': [], 'valid_acc': []}

        y_train_enc = self._onehot(y_train, n_output)

        # przebiegi po epokach uczenia
        for i in range(self.epochs):

            # przebiegi po mini-grupach
            indices = np.arange(X_train.shape[0])

            if self.shuffle:
                self.random.shuffle(indices)

            for start_idx in range(0, indices.shape[0] - self.minibatch_size +
                                   1, self.minibatch_size):
                batch_idx = indices[start_idx:start_idx + self.minibatch_size]

                # propagacja w przód
                z_h, a_h, z_out, a_out = self._forward(X_train[batch_idx])

                ##################
                # Propagacja wsteczna
                ##################

                # [n_próbek, n_etykiet_klas]
                sigma_out = a_out - y_train_enc[batch_idx]

                # [n_próbek, n_hidden]
                sigmoid_derivative_h = a_h * (1. - a_h)

                # [n_próbek, n_etykiet_klas] dot [n_ etykiet_klas, n_hidden]
                # -> [n_ próbek, n_hidden]
                sigma_h = (np.dot(sigma_out, self.w_out.T) *
                           sigmoid_derivative_h)

                # [n_cech, n_ próbek] dot [n_ próbek, n_hidden]
                # -> [n_cech, n_hidden]
                grad_w_h = np.dot(X_train[batch_idx].T, sigma_h)
                grad_b_h = np.sum(sigma_h, axis=0)

                # [n_hidden, n_ próbek] dot [n_ próbek, n_etykiet_klas]
                # -> [n_hidden, n_ etykiet_klas]
                grad_w_out = np.dot(a_h.T, sigma_out)
                grad_b_out = np.sum(sigma_out, axis=0)

                # Regularyzacja i aktualizowanie wag
                delta_w_h = (grad_w_h + self.l2*self.w_h)
                delta_b_h = grad_b_h  # obciążenie nie jest regularyzowane
                self.w_h -= self.eta * delta_w_h
                self.b_h -= self.eta * delta_b_h

                delta_w_out = (grad_w_out + self.l2*self.w_out)
                delta_b_out = grad_b_out  # obciążenie nie jest regularyzowane
                self.w_out -= self.eta * delta_w_out
                self.b_out -= self.eta * delta_b_out

            #############
            # Ewaluacja
            #############

            # Ewaluacja po każdej epoce w trakcie uczenia
            z_h, a_h, z_out, a_out = self._forward(X_train)
            cost = self._compute_cost(y_enc=y_train_enc,
                                      output=a_out)

            y_train_pred = self.predict(X_train)
            y_valid_pred = self.predict(X_valid)

            train_acc = ((np.sum(y_train == y_train_pred)).astype(np.float) /
                         X_train.shape[0])
            valid_acc = ((np.sum(y_valid == y_valid_pred)).astype(np.float) /
                         X_valid.shape[0])

            sys.stderr.write('\r%0*d/%d | Koszt: %.2f '
                             '| Dokładność uczenia/walidacji: %.2f%%/%.2f%% ' %
                             (epoch_strlen, i+1, self.epochs, cost,
                              train_acc*100, valid_acc*100))
            sys.stderr.flush()

            self.eval_['cost'].append(cost)
            self.eval_['train_acc'].append(train_acc)
            self.eval_['valid_acc'].append(valid_acc)

        return self
```

Wyuczymy teraz model MLP za pomocą 55 000 przykładów tworzących przetasowany zbiór
danych uczących MNIST, a pozostałe 5000 przykładów posłuży nam do walidacji.

Jak widać w powyższym listingu, wprowadziliśmy funkcję fit przyjmującą cztery argumenty
wejściowe: obrazy uczące, etykiety uczące, obrazy walidacyjne i etykiety walidacyjne. Podczas
uczenia sieci neuronowej naprawdę warto porównywać dokładność uczenia i walidacji, dzięki
czemu łatwiej jest nam ocenić skuteczność modelu przy uwzględnieniu architektury i hiperparametrów.
Na przykład, jeżeli zaobserwujemy małą dokładność uczenia i walidacji, to prawdopodobnie
istnieje jakiś problem z zestawem danych albo z konfiguracją hiperparametrów.
Względnie duża różnica pomiędzy dokładnością uczenia i walidacji może oznaczać przetrenowanie
modelu na zbiorze uczącym, przez co chcemy zmniejszyć liczbę parametrów modelu
lub zwiększyć siłę regularyzacji. Jeżeli model uzyskuje dużą dokładność zarówno podczas
uczenia, jak i walidacji, to model będzie prawdopodobnie dobrze uogólniał predykcje na podstawie
nowych danych, np. testowych, które wykorzystujemy do końcowej oceny modelu.

```python
n_epochs = 200

nn = NeuralNetMLP(n_hidden=100, 
                  l2=0.01, 
                  epochs=n_epochs, 
                  eta=0.0005,
                  minibatch_size=100, 
                  shuffle=True,
                  seed=1)

nn.fit(X_train=X_train[:55000], 
       y_train=y_train[:55000],
       X_valid=X_train[55000:],
       y_valid=y_train[55000:])
```

W implementacji kodu NeuralNetMLP zdefiniowaliśmy również atrybut eval_ gromadzący
wartość kosztu, dokładność uczenia i walidacji dla każdej epoki, dzięki czemu jesteśmy
w stanie zwizualizować wyniki w środowisku Matplotlib:

```python
import matplotlib.pyplot as plt


plt.plot(range(nn.epochs), nn.eval_['cost'])
plt.ylabel('Koszt')
plt.xlabel('Epoki')
#plt.savefig('rysunkis/12_07.png', dpi=300)
plt.show()
```

pozwolili na więcej przebiegów.

Przyjrzyjmy się teraz dokładności wobec zestawu uczącego i walidacyjnego:

```python
plt.plot(range(nn.epochs), nn.eval_['train_acc'], 
         label='Zestaw uczący')
plt.plot(range(nn.epochs), nn.eval_['valid_acc'], 
         label='Zestaw walidacyjny', linestyle='--')
plt.ylabel('Dokładność')
plt.xlabel('Epoki')
plt.legend(loc='lower right')
#plt.savefig('rysunki/12_08.png', dpi=300)
plt.show()
```

Z powyższego wykresu wnioskujemy, że odstęp pomiędzy dokładnością dla danych uczących
i walidacyjnych zwiększa się wraz z kolejnymi epokami. Mniej więcej w okolicach 50. epoki
dokładność dla zbiorów uczącego i walidacyjnego jest taka sama, po czym sieć stopniowo
staje się przetrenowana wobec danych uczących.
Zwróć uwagę, że przykład ten został wybrany celowo w celu ukazania zjawiska przetrenowania
i wyjaśnienia, dlaczego warto porównywać dokładność uczenia i walidacji w czasie trenowania
modelu. Jednym ze sposobów zmniejszenia przetrenowania jest zwiększenie siły regularyzacji —
np. wyznaczenie wartości l2 = 0.1.


Sprawdźmy jeszcze skuteczność uogólniania modelu poprzez wyliczenie dokładności przewidywań
wobec zestawu testowego:

```python
y_test_pred = nn.predict(X_test)
acc = (np.sum(y_test == y_test_pred)
       .astype(np.float) / X_test.shape[0])

print('Dokładność dla danych testowych: %.2f%%' % (acc * 100))
```

Pomimo nieznacznego przetrenowania wobec zbioru uczącego nasza względnie prosta sieć
neuronowa zawierająca jedną warstwę ukrytą uzyskała całkiem dobrą skuteczność dla zbioru
testowego, zbliżoną do dokładności wobec zestawu walidacyjnego (97,98%).


Sprawdźmy teraz, z jakimi cyframi ma problem omawiany model MLP. Pierwsza
liczba w opisie rysunków stanowi indeks analizowanego obrazu, druga wskazuje na rzeczywistą
etykietę klas (r), a trzecia przedstawia prognozowaną etykietę klas (p).
Jak widać, rozpoznanie niektórych z tych cyfr może stanowić wyzwanie nawet dla ludzkiego
mózgu. Przykładowo cyfra 6 pod numerem 8 wygląda jak niedbale napisana cyfra 0, natomiast
cyfra 8 pod numerem 23 mogłaby być dziewiątką, gdyż jej dolna część jest zwężona, a do tego napisana
z pogrubieniem.

```python
miscl_img = X_test[y_test != y_test_pred][:25]
correct_lab = y_test[y_test != y_test_pred][:25]
miscl_lab = y_test_pred[y_test != y_test_pred][:25]

fig, ax = plt.subplots(nrows=5, ncols=5, sharex=True, sharey=True)
ax = ax.flatten()
for i in range(25):
    img = miscl_img[i].reshape(28, 28)
    ax[i].imshow(img, cmap='Greys', interpolation='nearest')
    ax[i].set_title('%d) r: %d p: %d' % (i+1, correct_lab[i], miscl_lab[i]))

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
#plt.savefig('rysunki/12_09.png', dpi=300)
plt.show()
```
