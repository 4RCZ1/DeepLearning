## 1.3 Grafy obliczeniowe w TF1 i TF2 oraz wydajność ich kodów

w TF1 dane  trzeba było ulokować w <b>statycznym grafie</b> korzystając z <b>sesji</b> (Session). W TF2 dymyślnie działa mechanizm <b>wczesnego wykonania</b> (eager execution), a mechanizm sesji zniknął. Zaletą takiego rozwiązania jest uproszczenie kodu tworzenia modeli, natomiast wadą obniżenie wydajności (statystycznie razy 2 - ma na to wpływ również niezbyt doskonała integracja biblioteki keras jako modułu tensorflow).  

W ogólności zasadą koncepcyjną pisania kodu w TF2 jest jego podział na wiele małych funkcji, które mogą być one eksportowane i ponownie importowane w innych projektach. Ponadto funkcje takie   przez dekorator @tf.function mogą generować grafy statyczne, co ma zasaniczy wpływ na ich wydajność. 

Ponadto wskazane jest używanie zintegrowanego z TF2 modułu keras, który optymalizuje zarządzanie grupami zmiennych, pozwala je czasowo wyłączać z procesu uczenia (stan zamrożenia) oraz jeszcze bardziej kompaktuje kod.

Wczesne wykonanie kodu znaczenie upraszcza jego debugowanie, ale dubugowanie w TF2 jest również możliwe dla grafów statycznych (np. utworzonych przez @tf.function) przez funkcyjny przełącznik tf.config.run_functions_eagerly(True).

### 2.2. Koncepcje inicjalizowania zmiennych

W modelach sieci neuronowych inicjalizowanie parametrów modelu z losowymi
wagami jest niezbędne do złamania symetrii w fazie propagacji wstecznej, w przeciwnym wypadku wielowarstwowa sieć neuronowa nie byłaby lepsza od sieci jednowarstwowej, takiej jak
używana np. w regresji logistycznej. 

Podczas tworzenia obiektu Variable możemy wprowadzić
schemat inicjalizacji losowej. TensorFlow może generować liczby losowe za pomocą różnych rodzajów rozkładów; służy do tego klasa tf.random (więcej informacji znajdzieszna stronie [random]https://www.tensorflow.org/versions/r2.0/api_docs/python/tf/random). 

Poniżej  przyjrzymy się niektórym standardowym metodom inicjalizacji, które są również
dostępne w interfejsie Keras (zob. [keras initializers]https://www.tensorflow.org/versions/r2.0/api_docs/python/tf/
keras/initializers).

#### Inicjalizacja Xaviera (lub Glorota)

We wczesnym okresie rozwoju uczenia głębokiego zauważono, że inicjalizacja wag za pomocą losowego rozkładu
jednostajnego lub normalnego często prowadzi do słabej skuteczności modelu podczas uczenia.
W 2010 roku Xavier Glorot i Yoshua Bengio przyjrzeli się uważniej wpływowi inicjalizacji i zaproponowali
nowatorski, bardziej wszechstronny jej schemat, usprawniający uczenie sieci głębokich. Ogólny zamysł
inicjalizacji Xaviera polega na przybliżonym zrównoważeniu wariancji gradientów w różnych warstwach sieci.
Jeśli tego nie zrobimy, w trakcie uczenia uwaga może być zbytnio koncentrowana na jednej warstwie
kosztem innej warstwy.

Zgodnie z [artykułem](https://proceedings.mlr.press/v9/glorot10a/glorot10a.pdf),  jeżeli chcemy inicjalizować
wagi z rozkładem jednostajnym (uniform, uzn. <b>U</b>), powinniśmy wybrać następujący przedział tego rozkładu:

$$
W\sim U\left(-\frac{\sqrt{6}}{\sqrt{n_i+n_{i+1}}},\frac{\sqrt{6}}{\sqrt{n_{i}+n_{i+1}}}\right)
$$
gdzie $n_i$ oznacza ilość neuronów w i-tej warstwie.

Poniższy kod pokazuje  jak można zdefiniować
obiekt Variable wewnątrz klasy bazowej tf.Module. Zdefiniujemy dwie zmienne: modyfikowalną
i niemodyfikowalną:



```python
class MyModule(tf.Module):
    def __init__(self):
        init = tf.keras.initializers.GlorotNormal()
        self.w1 = tf.Variable(init(shape=(2, 3)), trainable=True)
        self.w2 = tf.Variable(init(shape=(1, 2)), trainable=False)
                
m = MyModule()
print('Wszystkie zmienne modułu: ', [v.shape for v in m.variables])
print('Zmienna modyfikowalna:   ', [v.shape for v in
                                 m.trainable_variables])
```

Jak widać na powyższym listingu, stworzenie podklasy klasy tf.Module daje nam bezpośredni
dostęp do wszystkich zmiennych zdefiniowanych w danym obiekcie (tutaj w wystąpieniu
naszej niestandardowej klasy MyModule) poprzez atrybut .variables.

Kolejny kod pokazuje jak można korzystać ze zmiennych wewnątrz funkcji dekorowanej za
pomocą dekoratora tf.function. Podczas definiowania obiektu Variable wewnątrz standardowej
(niedekorowanej) funkcji można oczekiwać, że nowy obiekt Variable będzie tworzony i inicjalizowany
wraz z każdym wywołaniem funkcji. Jednak dekorator tf.function będzie starał się
wykorzystywać istniejący już obiekt Variable na podstawie śledzenia i tworzenia grafu. Tensorflow uniemożliwia więc tworzenie obiektu Variable wewnątrz funkcji dekorowanej i w konsekwencji
poniższy listing będzie powodował błąd:


```python
## uruchomienie poniższego wiersza zakończy się komunikatem o błędzie
## ==> nie można tworzyć zmiennych wewnątrz funkcji dekorowanej

#@tf.function
#def f(x):
#    w = tf.Variable([1, 2, 3])

#f([1])

## ==> skutkuje błędem
## ValueError: tf.function-decorated function tried to create variables on non-first call.
```

Jednym ze sposobów uniknięcia tego problemu jest zdefiniowanie obiektu Variable poza funkcją dekorowaną i użycie go wewnątrz funkcji:

```python
import tensorflow as tf

tf.random.set_seed(1)
w = tf.Variable(tf.random.uniform((3, 3)))

@tf.function
def compute_z(x):    
    return tf.matmul(w, x)

x = tf.constant([[1], [2], [3]], dtype=tf.float32)
tf.print(compute_z(x))
```
